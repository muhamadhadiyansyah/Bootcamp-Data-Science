# Interquartile rank 

```python
# Menghitung interquartile rank pada kolom time_spend_company

# 1. mencari Q3
q3 = data['time_spend_company'].quantile(0.75)

# 2. mencari Q1
q1 = data['time_spend_company'].quantile(0.25)

# 3. menghitunga IQR 
IQR = q3 - q1
IQR
```

### BoxPlot

```python
# Membuat Boxplot pada kolom time_spend_company
plt.boxplot(data['time_spend_company'])

# Membuat judul
plt.title('Bloxplot Time Spend Company')

# Menampilkan visualisasi
plt.show()
```

Jika dilihat dari visualisasi boxplot, ternyata data time spend company memiliki 4 outlier. Maka jika ingin mengukur pemusatan jangan menggunakan mean tetapi menggunakan median. Selain itu, jika ingin mengukur penyebaran data, jangan menggunakan variance atau standar deviasi tetapi gunakan IQR. 

### Histogram

```python
# Membuat histogram dari kolom time_spend_company
plt.hist(data['time_spend_company'])

# Menunjukan nilai median dengan garis yang diberi warna kuning
# Caranya dengan menggunakan fungsi axvline
plt.axvline(data['time_spend_company'].median(),color='yellow')

# Menunjukan nilai mean dengan garis yang diberi warna hitam
plt.axvline(data['time_spend_company'].mean(),color='black')

# menampilkan hasil
plt.show()
```

Histogram menunjukan Positive Skewed yang mana terdapat nilai outliers atas. Selain itu dapat diperkuat dengan nilai mean yang lebih besar dari nilai median

# Correlation

```python
# Membuat scatter plot 
# x atau horizontal merupakan nilai pada kolom average monthly hours 
# y atau vertical satisfaction level, 
# c untuk melihat yang mana yang udah promosi diberi tanda
plt.scatter(x = data['average_montly_hours'], y = data['satisfaction_level'], c = data['promotion_last_5years'])

# Membuat x label atau horizontal label
plt.xlabel('Average Montly Hours')

# Membuat y label atau vertical label
plt.ylabel('Satisfaction Level')

# Membuat judul
plt.title('Relationship')

# menampilkan hasil
plt.show()
```

Scatter plot dapat digunakan untuk melihat korelasi. Dalam kasus ini, dapat diketahui bahwa tidak ada korelasi antara average_montly_hour dengan satisfaction_level. Karena visualisasi menunjukan data data yang acak atau tidak membentuk pola tertentu yang dapat diartikan bahwa tidak ada hubungan antar variabel.

```python
# Menghitung korelasi antara average_montly_hour dengan satisfaction_level
np.corrcoef(data['average_montly_hours'], data['satisfaction_level'])

# karena nilainya mendekati nol maka tidak berkorelasi
```

```python
# Kita juga dapat menggunakan fungsi korr untuk mengetahui korelasi dari keseluruhan kolom yang ada di dataset kita
data.corr()
# Jika kolom tersebut object string dianggap tidak sah atau tidak dapat diinterpretasi
```
