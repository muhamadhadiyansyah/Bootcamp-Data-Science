# Perbedaan Antara Matplotlib VS Seaborn

Visualisasi Data adalah representasi grafik dari data. Ini mengubah kumpulan data besar menjadi grafik kecil, sehingga membantu dalam analisis dan prediksi data. Ini adalah elemen yang sangat diperlukan dari ilmu data yang membuat data kompleks lebih mudah dipahami dan diakses. Matplotlib dan Seaborn bertindak sebagai tulang punggung visualisasi data melalui Python.

Matplotlib: Ini adalah pustaka Python yang digunakan untuk merencanakan grafik dengan bantuan pustaka lain seperti Numpy dan Pandas. Ini adalah alat yang ampuh untuk memvisualisasikan data dengan Python. Ini digunakan untuk membuat interferensi statis dan merencanakan grafik array 2D. Ini pertama kali diperkenalkan oleh John D. Hunter pada tahun 2002. Ini menggunakan Pyplot untuk menyediakan MATLAB seperti antarmuka gratis dan open-source. Ia mampu menangani berbagai sistem operasi dan backend grafisnya.

Seaborn : Ini juga merupakan pustaka Python yang digunakan untuk merencanakan grafik dengan bantuan Matplotlib, Pandas, dan Numpy. Itu dibangun di atas atap Matplotlib dan dianggap sebagai superset dari perpustakaan Matplotlib. Ini membantu dalam memvisualisasikan data univariat dan bivariat. Ini menggunakan tema yang indah untuk mendekorasi grafik Matplotlib. Ini bertindak sebagai alat penting dalam menggambarkan Model Regresi Linier. Ini berfungsi dalam membuat grafik data Time-Series statis. Ini menghilangkan tumpang tindih grafik dan juga membantu mempercantiknya.

```python
import seaborn as sns
import numpy as np
from numpy import median
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Bar Plot
sns.barplot(x = 'day', y = 'tip', data = tips)
```

```python
# untuk menghilangkan tulisan <AxesSubplot:xlabel='day', ylabel='tip'>, pake titik koma di akhir code
sns.barplot(x = 'day', y = 'tip', data = tips);
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips);
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex');
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', palette = 'winter_r');
# palette untuk memilih warna yang diinginkan, hue untuk pembagi, data untuk mengambil data yang ingin divisualisasikan
```

```python
# berdasarkan smokar
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker');
```

```python
# ingin menampilkan secara horizontal
sns.barplot(x = 'total_bill', y = 'day', data = tips, palette = 'spring');
```

```python
# order
sns.barplot(x = 'day', y = 'tip', data = tips, palette = 'spring', order = ['Sat', 'Fri', 'Sun', 'Thur']);
```

```python
# estimator median
sns.barplot(x = 'day', y = 'total_bill', data = tips, estimator = median, palette = 'spring');
```

```python
sns.barplot(x = 'smoker', y = 'tip', data = tips, estimator = median, palette = 'coolwarm');
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, palette = 'spring', capsize = 0.3);
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, palette = 'spring', capsize = 0.9);
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', palette = 'husl', capsize = 0.1);
```

```python
sns.barplot(x = 'day', y = 'total_bill', data = tips, palette = 'autumn');
```

```python
# Distribution plot
num = np.random.randn(150)
sns.distplot(num);
```

```python
sns.distplot(num, color = 'red');
```

```python
label_dist = pd.Series(num, name = 'variabel x')
sns.distplot(label_dist);
```

```python
sns.distplot(label_dist, vertical = True);
```

```python
sns.distplot(label_dist, hist = False)
```

```python
sns.distplot(label_dist, rug = True, hist = False)
```
