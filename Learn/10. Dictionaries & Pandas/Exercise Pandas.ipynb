## Load / Create Data Frame
We can read CSV or excel files to import data. We can also create dataframes programatically like this:

```python
df = pd.DataFrame({'Artist':['Billie Holiday','Jimi Hendrix', 'Miles Davis', 'SIA'],
              'Genre': ['Jazz', 'Rock', 'Jazz', 'Pop'],
              'Listeners': [1300000, 2700000, 1500000, 2000000],
              'Plays': [27000000, 70000000, 48000000, 74000000]})
```

 The variable `df` now contains a dataframe. Executing this cell shows the contents of the dataframe

```python
df
```

## Selection
We can select any column using its label:

```python
df['Artist']
```

We can select one or multiple rows using their numbers (inclusive of both bounding row numbers):

```python
df.loc[1:3]
```

```python
# Exercise: Select only row #0
df.loc[0]
```

We can select any slice of the table using a both column label and row numbers using loc:

```python
df.loc[1:3,['Artist']]
```

```python
# Exercise: Select 'Artist' and 'Plays' for rows #1 and #2
df.loc[1:2, ['Artist', 'Plays']]
```

## Filtering
Now it gets more interesting. We can easily filter rows using the values of a specific row. For example, here are our jazz musicians:

```python
df[df['Genre'] == "Jazz" ]
```

```python
# Exercise: Select the row where the Genre is 'Rock'
df[df['Genre'] == "Rock" ]
```

Here are the artists who have more than 1,800,000 listeners:

```python
df[df['Listeners'] > 1800000 ]
```

```python
# Exercise: Select the rows where 'Plays' is less than 50,000,000
df[df['Listeners'] < 50000000 ]
```

# Ex1 - Filtering and Sorting Data


### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'

chipo = pd.read_csv(url, sep = '\t')
```

### Step 4. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
# select only the item_name and item_price columns
price_per_item = chipo[['item_name', 'item_price']]
price_per_item
```

### Step 5. How many times was a Veggie Salad Bowl ordered?

```python
chipo_salad = chipo[chipo.item_name == "Veggie Salad Bowl"]

len(chipo_salad)
```

### Step 6. How many times did someone order more than one Canned Soda?

```python
chipo_drink_steak_bowl = chipo[(chipo.item_name == "Canned Soda") & (chipo.quantity > 1)]
len(chipo_drink_steak_bowl)
```

# Ex2 - Filtering and Sorting Data

This time we are going to pull data directly from the internet.

### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv). 

### Step 3. Assign it to a variable called euro12.

```python
euro12 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv', sep=',')
euro12
```

### Step 4. Select only the Goal column.

```python
euro12.Goals
```

### Step 5. How many team participated in the Euro2012?

```python
euro12.shape[0]
```

### Step 6. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

```python
# filter only giving the column names

discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
discipline
```

### Step 7. Sort the teams by Red Cards, then to Yellow Cards

```python
discipline.sort_values(['Red Cards', 'Yellow Cards'], ascending = False)
```

### Step 8. Calculate the mean Yellow Cards given per Team

```python
round(discipline['Yellow Cards'].mean())
```

### Step 9. Filter teams that scored more than 6 goals

```python
euro12[euro12.Goals > 6]
```

### Step 10. Select the teams that start with G

```python
euro12[euro12.Team.str.startswith('G')]
```

### Step 11. Select the first 7 columns

```python
# use .iloc to slices via the position of the passed integers
# : means all, 0:7 means from 0 to 7

euro12.iloc[: , 0:7]
```

### Step 12. Select all columns except the last 3.

```python
# use negative to exclude the last 3 columns

euro12.iloc[: , :-3]
```

### Step 13. Present only the Shooting Accuracy from England, Italy and Russia

```python
# .loc is another way to slice, using the labels of the columns and indexes

euro12.loc[euro12.Team.isin(['England', 'Italy', 'Russia']), ['Team','Shooting Accuracy']]
```
