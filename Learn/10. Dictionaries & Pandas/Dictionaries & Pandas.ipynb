# Understanding Dictionaries

Dictionary adalah tipe data yang anggotanya terdiri dari pasangan kunci:nilai (key:value). 

Dictionary bersifat tidak berurut (unordered) sehingga anggotanya tidak memiliki indeks.

image.png

## Creating Dictionary

```python
# Membuat dictionary kosong
mydict = {}
mydict
```

```python
# dictionary dengan kunci integer
mydict = {1: 'Sepatu', 2: 'Tas'}
mydict
```

```python
# dictionary dengan kunci campuran
mydict = {'warna':'merah', 1:[2, 3, 5]}
mydict
```

```python
# membuat dictionary menggunakan fungsi dict()
my_dict = dict([(1, 'sepatu'), (2, 'bola')])
my_dict
```

## Accesing Dictionary

```python
# get vs [] for retrieving elements
mydict = {'name': 'Jack', 'age':26}
```

```python
# Output: Jack
print(mydict['name'])
```

```python
# Output: 26
print(mydict.get('age'))
```

```python
# Trying to access keys which doesn't exist throws error
# Output None
print(my_dict.get('address'))
```

```python
# KeyError
print(my_dict['address', 'age'])
```

## Dictionary Manipulation

```python
# Changing and adding Dictionary Elements
my_dict = {'name': 'Jack', 'age':26}
```

```python
# Update value
my_dict['age'] = 27
```

```python
# Output: {'age': 27, 'name': 'Jack'}
print(my_dict)
```

```python
# add item
my_dict['address'] = 'Downtown'
```

```python
# Output: {'address': 'Downtown', 'age': 27, 'name': 'Jack'}
print(my_dict)
```

```python
a_dict = {'a':1, 'b':2, 'c':3}
```

```python
new_key = 'A'
old_key = 'a'
```

```python
a_dict[new_key] = a_dict.pop(old_key)
```

```python
print(a_dict)
```

## Removing Dictionary

```python
# Removing elements from a dictionary

# Create a dictionary
squares = {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}
```

```python
# Remove a particular item, returns its value
# Output: 16
print(squares.pop(4))

# Output: {1: 1, 2: 4, 3: 9, 5: 25}
print(squares)
```

```python
# remove an arbitrary item, return (key,value)
# Output: (5, 25)
print(squares.popitem())
# Output: {1: 1, 2: 4, 3: 9}
```

```python
# remove all items
print(squares)

# Output: {}
del squares
```

```python
# delete the dictionary itself
print(squares)
# Throws Error
```

## Create a DataFrame from Dict of ndarrays / Lists

```python
import pandas as pd
data = {'Name': ['Tom', 'Jack', 'Steve', 'Ricky'], 'Age': [28, 24, 29, 27]}
df = pd.DataFrame(data)
print(df)
```

## Create a DataFrame from List of Dicts

```python
#With two column indices, values same as dictionary keys
data = [{'a' : 1, 'b' : 2}, {'a': 5, 'b': 10, 'c': 20}]

#With two column indices with one index with other name
df1 = pd.DataFrame(data, index = ['first', 'second'], columns = ['a', 'b', 'c'])
print(df1)
```

## Selecting, Slicing and Filtering data in a Pandas DataFrame

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.arange(20).reshape(5,4), columns=["A","B","C","D"])
df
```

```python
# select one column
df.loc[:, 'A']
```

```python
# select one column
df['A']
```

```python
# select one column
df.A
```

```python
# select multiple columns
df.loc[:,['A', 'C']]
```

```python
df[['A', 'C']]
```

```python
# Select a row by its label.
df.loc[0]
```

```python
# Select multiple rows by label
df.loc[[0, 1]]
```

```python
# Accessing values by row and column label.
df.loc[0, 'D']
```

```python
# Accessing values from multiple columns of same row.
df.loc[1, ['A', 'C']]
```

```python
# Accessing values from multiple rows but same column.
df.loc[[0, 1], 'B']
```

## Select by Index Position

```python
# Select a row by index location.
df.iloc[0]
```

```python
df
```

```python
# Select a column by index location.
df.iloc[:, 3]
```

```python
# Select data at the specified row and column location.
df.iloc[0, 3]
```

```python
#Slice columns by label.
df.loc[:, 'B':'D']
```

```python
# To slice columns by index position.
df.iloc[0,:]
```

```python
# To slice row and columns by index position.
df.iloc[0:2, 0:2]
```
