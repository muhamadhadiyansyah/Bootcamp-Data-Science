## Clustering Analysis

```python
# Mengelompokan customer berdasarkan variabel umur, pendapatan, dan spending score
df_cluster = data.iloc[:, [2, 3, 4]] 
df_cluster
```

## Kmeans Algorithm

**The Elbow Method to find the No. of Optimal Clusters**

Menentukan nilai K

```python
from sklearn.cluster import KMeans

# Dalam melakukan tunning, mencoba satu-persatu dengan looping
wcss = [] # nantinya akan berisikan nilai Inertia dari masing-masing nilai K
for i in range(1, 11):
    km = KMeans(n_clusters = i, init = 'k-means++', random_state = 123) # init: initial centroid
    km.fit(df_cluster)
    wcss.append(km.inertia_)

# Membuat plot sumbu x adalah range 1-10, sumbu y adalah wcss    
plt.figure(figsize=(15,5))
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

Berdasarkan plot, kita melihat kondisi ketika penurunan tidak drastis untuk menentukan jumlah cluster. Dari kasus ini, kita memilih 5 cluster.

**Visualizaing the Clusters**

```python
# Membuat model K-Means
km = KMeans(n_clusters = 5, init = 'k-means++', random_state = 123)
cluster = km.fit_predict(df_cluster)
```

```python
# Melihat hasil prediksi
cluster
```

```python
# Data hasil cluster dimasukan ke dalam data awal dengan nama kolom cluster
df_cluster['cluster'] = cluster
```

```python
df_cluster
```

```python
# Untuk mengetahui karakteristik dari masing-masing cluster
hasil1 = df_cluster.groupby(['cluster']).agg({"Age":"mean", "Annual Income (k$)":"mean", "Spending Score (1-100)":"mean"}).reset_index() # reset index agar index urut
hasil1
```

## Hierarchial Clustering

**Using Dendrograms to find the no. of Optimal Clusters**

```python
# Membuat Dendrogram agar jumlah clusternya jelas
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(15,10))
dendrogram = sch.dendrogram(sch.linkage(df_cluster.drop(labels=['cluster'], axis=1), method = 'ward')) # drop kolom cluster karena hasil dari k-means
plt.title('Dendrogam', fontsize = 20)
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

Dari dendogram dapat dapat diketahui jumlah cluster optimum 3. Berdasarkan 3 warna. Agar hasilnya sama seperti K-Means, kita mencoba 5 cluster

**Visualizing the Clusters of Hierarchial Clustering**

```python
# Membuat Hierarchial Clustering
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 5, affinity = 'euclidean', linkage = 'ward')
cluster_hc = hc.fit_predict(df_cluster.drop(labels=['cluster'], axis=1))
```

```python
# Hasil prediksi
cluster_hc
```

```python
# Memasukan ke dalam data, dengan kolom cluster_hc
df_cluster['cluster_hc'] = cluster_hc
```

```python
df_cluster
```

```python
# Melihat karakteristik dari masing-masing cluster
hasil2 = df_cluster.groupby(['cluster_hc']).agg({"Age":"mean","Annual Income (k$)":"mean","Spending Score (1-100)":"mean"}).reset_index()
hasil2
```
