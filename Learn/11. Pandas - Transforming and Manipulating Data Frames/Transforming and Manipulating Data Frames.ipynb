## Pandas DataFrame Sorted by Values

DataFrame.sort_values(by, axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None)

Return Type is DataFrame or None.
If sorted inplace return type is None, otherwise DataFrame.

```python
import pandas as pd
df = pd.read_csv("data1.csv") 
df
```

### Sorting values by one column "EmpID"

```python
df.sort_values('EmpID')
```

### Sorting values by one column "EmpID" in descending order

```python
df.sort_values('EmpID', ascending=False)
```

### Sorting values based on two columns "Skill" and "EmpID"

```python
df.sort_values(by=['Skill','EmpID'])
```

### Sorting dataframe by one column "Skill"

```python
df.sort_values(by=['Skill'])
```


### Sorting dataframe by two column in two different order

```python
df.sort_values(by=['Skill', 'EmpID'], ascending=[True, False])
```

### Sorting dataframe by putting NaN values first

```python
df.sort_values("Salary", na_position = 'first')
```

### Sorting in place

```python
df.sort_values("Salary", inplace = True)
df
```

### Sorting by ignoring index


```python
df.sort_values("Salary", ignore_index = True)
```

### Sorting dataframe by applying key function

```python
df.sort_index(inplace = True)
```

```python
df.sort_values(by=["Skill"], key=lambda x:x.str.lower())
```

```python
df.sort_values(by=["Skill"], key=lambda x:x.str.len())
```

# Data Description
In these exercises you'll use a data that contains information for tips.

Details for the columns:
* **total_bill**: a numeric vector, **the bill amount (dollars)**
* **tip**: a numeric vector, **the tip amount (dollars)**
* **sex**: a factor with levels **\[Female, Male\], gender of the payer of the bill**
* **smoker**: a factor with levels **\[No, Yes\], whether the party included smokers**
* **day**: a factor with levels **\['Sun', 'Sat', 'Thur', 'Fri'\], day of the week**
* **time**: a factor with levels **\[Dinner, Launch\], rough time of day**
* **size**: a numeric vector, **number of people in party**

Get all records where the **day is Saturday or Sunday**. Remember that **the values in the day column** are written like that: **'Sun', 'Sat', 'Thur', 'Fri'**

## Task 7
Get all records where **the number of people is greater than 4** and **the payer of the bill** (the sex column) **is a female**.

```python
tips_data[(tips_data["size"] > 4) & (tips_data.sex == "Female")]
```

## Task 8
Get all records where **the total bill is greater than 40** and **the tip is greater than 5**.

```python
tips_data[(tips_data.total_bill > 40) & (tips_data.tip > 5)]
```

## Task 9
Get all records where the number of **the people is greater than 4** or **it's a dinner** party (the time column).

```python
tips_data[(tips_data["size"] > 4) | (tips_data.time == "Dinner")].head()
```

## Task 10
Get all records where the **the party doesn't include smokers**, the **total bill is greater than 30**, and the **day is Thurday or Friday**.

```python
smoker_filter = tips_data.smoker == "No"
total_bill_filter = tips_data.total_bill > 30
day_filter = tips_data.day.isin(["Thur", "Fri"])
all_filters = smoker_filter & total_bill_filter & day_filter

tips_data[all_filters]
```

## Task 11
Get all records where the **day is Friday or Saturday**, the **time is a lunch**, and the **number of the people is less than 4**.

```python
day_filter = tips_data.day.isin(["Fri", "Sat"])
time_filter = tips_data.time == "Lunch"
size_filter = tips_data["size"] < 4
all_filters = day_filter & time_filter & size_filter

tips_data[all_filters]
```

## Task 12

Get all records where 

the **day is Friday**, the **number of the people is less than 3** and the **payer of the bill is a female** 

or 

the **day is Sunday**, the **number of the people is greater than 4** and the **payer of the bill is a male**

```python
first_filter = (tips_data.day == "Fri") & (tips_data["size"] < 3) & (tips_data.sex == "Female")
second_filter = (tips_data.day == "Sun") & (tips_data["size"] > 4) & (tips_data.sex == "Male")
all_filters = first_filter | second_filter

tips_data[all_filters]
```

# Create a new column in Pandas DataFrame based on the existing columns

```python
# importing pandas as pd 
import pandas as pd 
  
# Creating the DataFrame 
df = pd.DataFrame({'Date':['10/2/2011', '11/2/2011', '12/2/2011', '13/2/2011'], 
                    'Event':['Music', 'Poetry', 'Theatre', 'Comedy'], 
                    'Cost':[10000, 5000, 15000, 2000]}) 
  
# Print the dataframe 
print(df) 
```

```python
# using apply function to create a new column 
df['Discounted_Price'] = df.apply(lambda row: row.Cost - 
                                  (row.Cost * 0.1), axis = 1) 
  
# Print the DataFrame after addition 
# of new column 
print(df) 
```

# Pandas DataFrame.transform

```python
# importing pandas as pd 
import pandas as pd 

# Creating the DataFrame 
df = pd.DataFrame({"A":[12, 4, 5, None, 1], 
                "B":[7, 2, 54, 3, None], 
                "C":[20, 16, 11, 3, 8], 
                "D":[14, 3, None, 2, 6]}) 

# Create the index 
index_ = ['Row_1', 'Row_2', 'Row_3', 'Row_4', 'Row_5'] 

# Set the index 
df.index = index_ 

# Print the DataFrame 
print(df)
```

```python
# add 10 to each element of the dataframe 
result = df.transform(func = lambda x : x + 10) 

# Print the result 
print(result) 
```

# Reindexing in Pandas DataFrame

Reindexing the Rows

```python
# import numpy and pandas module 
import pandas as pd 
import numpy as np 

column=['a','b','c','d','e'] 
index=['A','B','C','D','E'] 

# create a dataframe of random values of array 
df1 = pd.DataFrame(np.random.rand(5,5), 
            columns=column, index=index) 

print(df1) 

print('\n\nDataframe after reindexing rows: \n', 
df1.reindex(['B', 'D', 'A', 'C', 'E'])) 
```

```python
# import numpy and pandas module 
import pandas as pd 
import numpy as np 

column = ['a', 'b', 'c', 'd', 'e'] 
index = ['A', 'B', 'C', 'D', 'E'] 

# create a dataframe of random values of array 
df1 = pd.DataFrame(np.random.rand(5, 5), 
        columns = column, index = index) 

# create the new index for rows 
new_index =['U', 'A', 'B', 'C', 'Z'] 

print(df1.reindex(new_index)) 
```

Reindexing the columns using axis keyword

# DataFrame

```python
import numpy as np
df_1 = pd.DataFrame(np.arange(1,17).reshape(4,4),
                    index= ['Fi', 'Se', 'Th', 'Fo'],
                    columns = ['a', 'b', 'c', 'd'])

df_2 = pd.DataFrame(np.arange(1,17).reshape(4,4) * 10,
                    index= ['Fi', 'Se', 'Th', 'Fo'],
                    columns = ['a', 'b', 'c', 'd'])
```

```python
# Addition of two dataframes with the same index
df_1 + df_2
```

```python
# Multiplication of two dataframes with the same index
df_1 * df_2
```
