# Arithmetic Operators

Operator merupakan suatu simbol dalam Python yang digunakan untuk menjalankan perhitungan aritmatika atau logika. 

Nilai yang dioperasikan oleh operator disebut operan.

```python
# Contoh
5 + 6

# simbol + merupakan operator (penjumlahan)
# angka 5 dan 6 merupakan operan
# angka 11 merupakan hasil dari operasi
```

**Macam-macam operator aritmatika**
*   "+"  : Penjumlahan
*   "-"  : Pengurangan
*   "*"  : Perkalian
*   /  : Pembagian
*   %  : Modulo
*   //  : Floor Division
*   **  : Eksponen

Catatan:

Modulo

Ketika 7 dibagi 4 itu sama dengan 1Â¾. Dengan modulo maka 7 % 4 = 3 (merupakan sisa dari pembagian 7:4) Contoh lainnya:

11 % 4 = 3 (11 sama dengan 4 dikali 2 ditambah 3)

8 % 2 = 0 (8 sama dengan 2 dikali 4 ditambah 0)

133 % 12 = 1 (132 sama dengan 12 dikali 11 ditambah 1)

> Floor Divison

hasil dari perhitungan pembagian yang dibulatkan ke nilai terbawah. misal 12 : 5 maka hasilnya 2 koma. Dengan floor division 
dibulatkan menjadi 2.

> Eksponen

Eksponen merupakan perpangkatan dalam python. misalnya 2 pangkat 3 maka cara menghitungnya 2**3

```python
# Contoh
x = 18
y = 11

# Output Penjumlahan
print('Penjumlahan: ' + 'x + y =', x + y)

# Output Pengurangan
print('Pengurangan: ' + 'x - y =', x - y)

# Output Perkalian
print('Perkalian: ' + 'x * y =', x * y)

# Output Pembagian
print('Pembagian: ' + 'x / y =', x / y)

# Output Modulo
print('Modulo: ' + 'x % y =', x % y)

# Output Floor Divison
print('Floor Divison: ' + 'x // y =', x // y)

# Output eksponen
print('Eksponensial: ' + 'x ** y =', x ** y)
```

# Logical Operators

Contoh "AND", terdapat dua perhitungan, A dan B. 

A menampilkan hasil dari comparasion operator yang mana hasilnya True, dan B menampilkan hasil dari comparasion operator yang mana hasilnya True. 

Ketika A dan B digabungkan dengan logical operators maka hasilnya True. Namun jika salah satu antara A atau B menghasilkan output False, maka hasilnya False. 

Keduanya harus True agar menghasilkan nilai True. Sedangkan untuk "OR", Keduanya tidak harus True untuk menghasilkan output True. Apabila hasil A itu True dan B itu False, maka logical OR akan menghasilkan output True.

```python
x = True
y = False

print('x and y is', x and y)
print('x or y is', x or y)
print('not x is', not x)
```

# Introduction for Loops

Perulangan for di Python digunakan untuk mengulangi urutan (list, tuple, string) atau objek iterable lainnya. 

Iterasi pada suatu urutan disebut traversal.

Val adalah variabel yang mengambil nilai item di dalam urutan pada setiap iterasi.

Pengulangan berlanjut hingga kita mencapai item terakhir dalam urutan. Body for loop dipisahkan dari kode lainnya menggunakan indentasi.

Alur Kerja For Loops:


```python
# Program to find the sum of all numbers stored in a list

# List of numbers
numbers = [6, 5, 3, 8, 4, 2, 5, 4, 11]

# variable to store the sum
sum = 0

# iterate over the list
for val in numbers:
  sum = sum + val

print("The sum is", sum)
```

```python
Total_skor = [75, 39, 48, 69, 30, 60, 90]
jumlah_total_skor = 0
for val in Total_skor:
  jumlah_total_skor += val

print('Total skor kamu adalah', jumlah_total_skor)
```

#  The Range Function

Kita bisa menghasilkan urutan angka menggunakan fungsi range (). 

range (10) akan menghasilkan angka dari 0 sampai 9 (10 angka).

Dapat menggunakan start, stop dan step sebagai range (start, stop, step_size). 

step_size default ke 1 jika tidak disediakan.

Objek range adalah "malas" dalam arti karena tidak menghasilkan setiap angka yang "dikandungnya" saat kita membuatnya.

Dikarenakan ini bukan merupakan iterator, maka fungsi ini dapat mendukung operasi in, len dll

Fungsi ini tidak menyimpan semua nilai dalam memori; itu akan menjadi tidak efisien. Jadi, ia mengingat ukuran mulai, berhenti, langkah, dan menghasilkan angka berikutnya saat dalam pengoperasian.

Untuk memaksa fungsi ini mengeluarkan semua item, kita dapat menggunakan function list ().

```python
# Contoh The Range Function
print(range(10))
print(list(range(10)))
```

```python
# contoh menggunakan start:stop
# misal 2:8
print(list(range(2, 8)))

# misal 3 : 13
print(list(range(3, 13)))
```

```python
# contoh menggunakan start : stop : Step
# misal 2 : 20 : 3
print(list(range(2, 20, 3)))

# misal 1 : 44 : 2
print(list(range(1, 44, 2)))
```

# Introduction while loops

Perulangan while dalam Python digunakan untuk mengulang satu blok kode selama ekspresi uji (kondisi) benar. 

Biasanya menggunakan perulangan ini ketika kami tidak mengetahui berapa kali untuk mengulang sebelumnya.

Alur kerja While loop:


```python
# program to add natural
# numbers up to
# sum = 1 + 2 + 3 + ... + n


n = 10

# initialize sum and counter
sum = 0
i = 1

while i <= n:
  sum = sum + i
  i = i + 1

# print the sum
print('The sum is', sum)
```

# Break Statements (While Loops)

Pernyataan break menghentikan loop yang berisi itu. Kontrol program mengalir ke pernyataan segera setelah badan loop.

Jika pernyataan break berada di dalam loop nested (loop di dalam loop lain), pernyataan break akan menghentikan loop terdalam.

```python
# use of break statement inside the loop

for val in "string":
  if val == "i":
    break
  print(val)

print("The end")
```

```python
for val in list(range(10)):
  if val == 7:
    break
  print(val)

print("Stop")
```

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

# Continue Statements (While Loops)

Pernyataan continue digunakan untuk melewati sisa kode di dalam loop hanya untuk iterasi saat ini. 

Loop tidak berhenti tetapi berlanjut dengan iterasi berikutnya.

```python
# Program to show the use of continue statement inside loops

for val in 'string':
  if val == 'n':
    continue
  print(val)

print("The end")
```

```python
nilai_mahasiswa = [77, 89, 35, 76, 87, 90, 50, 77]
print("Daftar Peserta yang Lolos Seleksi CPNS")

for val in nilai_mahasiswa:
  if val <= 75:
    continue
  print(val)

print("Terima kasih")
```

# Loop Dictonary

Anda dapat melakukan loop melalui dictionary dengan menggunakan for loop.

Saat mengulang melalui dictionary, nilai yang dikembalikan adalah kunci dari dictionary, tetapi ada metode untuk mengembalikan nilai juga.

```python
thisdict = {
    'brand' : 'Ford',
    'model' : 'mustang',
    'year' : 1998
}
print(thisdict)
```
